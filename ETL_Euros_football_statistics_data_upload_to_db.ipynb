```python
def create_database():
    conn = psycopg2.connect("host = localhost dbname  = postgres user = postgres password = root")
    cur = conn.cursor()
    conn.set_session(autocommit = True)

    cur.execute("DROP DATABASE IF EXISTS euros_results ")
    cur.execute("CREATE DATABASE euros_results")

    
    conn.close()

    conn = psycopg2.connect("host = localhost dbname = euros_results user = postgres password = root")
    cur = conn.cursor()
    conn.set_session(autocommit = True)

    return conn, cur
```

```python
conn, cur = create_database()
```

```python
df_coaches.head()
```

```python
cur.execute("CREATE TABLE IF NOT EXISTS COACHES (Country_code VARCHAR PRIMARY KEY,\
             Country VARCHAR, Name VARCHAR, Role VARCHAR, id_match NUMERIC, Year NUMERIC)")
conn.commit()
```

```python
insert_coaches = """
    INSERT INTO COACHES(Country_code, Country, Name, Role, id_match, Year)
    VALUES(%s, %s, %s, %s, %s, %s)
    ON CONFLICT (Country_code) DO UPDATE SET
        Country = EXCLUDED.Country,
        Name = EXCLUDED.Name,
        Role = EXCLUDED.Role,
        id_match = EXCLUDED.id_match,
        Year = EXCLUDED.Year
"""

for i, row in df_coaches.iterrows():
    value = row.tolist()
    cur.execute(insert_coaches, value)
    #print(row.tolist())
   
#conn.commit()
```

```python
insert_lineups = """INSERT INTO LINEUPS(
                Country_code,
                Name ,
                Jersey_number,
                DOB ,
                id_match ,
                Year )VALUES
                (%s,%s,%s,%s,%s,%s)"""

for i, row in df_lineups.iterrows():
    values = row.tolist()
    cur.execute(insert_lineups, values)
```
